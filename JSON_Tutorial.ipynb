```python
import json

order_summary =[]

for customer in data:
  customer_orders = customer.get('orders', [])

  for order in customer_orders:
    oid = order.get('order_id')
    amount = order.get('total')

    order_summary.append((oid, amount))

for item in order_summary:
  print(f"Order ID: {item[0]} | Amount: {item[1]}")
```

```python
import pandas as pd
import json

with open('/content/sample.json', 'r') as file:
  data = json.load(file)


df_flat = pd.json_normalize(
    data,
    record_path=['orders', 'items'],
    meta=['customer_id', ['orders', 'order_id'], ['orders', 'total']]
)

df_flat.columns = ['Product', 'Price', 'Quantity', 'Customer ID', 'Order ID', 'Order Total']

df_flat = df_flat[['Customer ID', 'Order ID', 'Product', 'Quantity', 'Order Total']]

df_flat
```
