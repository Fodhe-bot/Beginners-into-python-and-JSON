```python
from PyPDF2 import PdfReader

# Load the PDF file
# Change the address to your file
reader = PdfReader("sample-sdf-document.pdf")

# Check the total number of pages
print(f"The PDF has {len(reader.pages)} pages.")
```

```python
import fitz

doc = fitz.open("sample-sdf-document.pdf")

print("PDF Metadata:", doc.metadata)
print(f"\nTotal pages:  {len(doc)}")
```

```python
page =doc[0]
print(page.get_text())
```

```python
page =doc[1]
print(page.get_text())
```

```python
page =doc[2]
print(page.get_text())
```

```python
text_data = page.get_text("dict")

for block in text_data.get("blocks", []):
    if "lines" in block:
        for line in block["lines"]:
            for span in line.get("spans", []):
                # Now you have font info too!
                font_size = span["size"]
                text = span["text"].strip()
                bbox = span["bbox"]

                if text: # Skip empty strings
                    print(f"[{font_size}] Text: {text} | Bbox: {bbox}")
```
