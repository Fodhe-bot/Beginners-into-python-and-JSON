```python
# Convert CCAvg to numeric, turning errors into NaNs
df['CCAvg'] = pd.to_numeric(df['CCAvg'], errors='coerce')

# Check if any NaNs were created by the conversion
print(f"Missing values in CCAvg: {df['CCAvg'].isnull().sum()}")
```

```python
# 1. Force the column to be strings again
# 2. Replace the slash with a dot
# 3. Convert to numeric properly
df['CCAvg'] = df['CCAvg'].astype(str).str.replace('/', '.').astype(float)

# 4. Check the results
print(df['CCAvg'].head())
print(f"Missing values now: {df['CCAvg'].isnull().sum()}")
```

```python
# 1. Reload (Make sure to use your actual file name)
df = pd.read_csv('/content/bank_loan.csv')

# 2. Clean the 'Experience' first (since we know it's a problem)
df.loc[df['Experience'] < 0, 'Experience'] = 0

# 3. Clean the 'CCAvg' slashes while they are still strings
df['CCAvg'] = df['CCAvg'].str.replace('/', '.')

# 4. NOW convert to numeric
df['CCAvg'] = pd.to_numeric(df['CCAvg'], errors='coerce')

# 5. Final check
print(df['CCAvg'].head())
print(f"Success! Missing values: {df['CCAvg'].isnull().sum()}")
```

```python
# 1. Convert to string and pad with zeros to ensure 5 digits
df['ZIP Code'] = df['ZIP Code'].astype(str).str.zfill(5)

# 2. Check a few to make sure it worked
print(df['ZIP Code'].head())

# 3. Check the type (it should now say 'object')
print(df['ZIP Code'].dtype)
```
